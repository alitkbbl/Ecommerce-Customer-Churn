```python
# ═══════════════════════════════════════════════════════════════════════════════
# NOTEBOOK 03: UNIVARIATE EXPLORATORY DATA ANALYSIS
# Purpose: Individual feature analysis and distribution exploration
# ═══════════════════════════════════════════════════════════════════════════════

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")
pd.set_option('display.max_columns', None)

# ═══════════════════════════════════════════════════════════════════════════════
# 1. LOAD DATA
# ═══════════════════════════════════════════════════════════════════════════════

df = pd.read_csv('../data/raw/customer_churn_dataset.csv')

print("="*80)
print("UNIVARIATE EXPLORATORY DATA ANALYSIS")
print("="*80)
print(f"\nDataset: {df.shape[0]:,} rows × {df.shape[1]} columns\n")

# ═══════════════════════════════════════════════════════════════════════════════
# 2. NUMERICAL FEATURES ANALYSIS
# ═══════════════════════════════════════════════════════════════════════════════

numerical_cols = ['Age', 'TenureMonths', 'DaysSinceLastInteraction', 'MonthlyCharges', 
                  'TotalCharges', 'NumProducts', 'SupportCalls', 'SatisfactionScore', 
                  'AvgMonthlyUsageGB', 'LatePayments']

print("="*80)
print("NUMERICAL FEATURES ANALYSIS")
print("="*80)

# Complete descriptive statistics
stats_df = df[numerical_cols].describe(percentiles=[.01, .05, .25, .5, .75, .95, .99]).T
stats_df['skewness'] = df[numerical_cols].skew()
stats_df['kurtosis'] = df[numerical_cols].kurtosis()
stats_df['cv'] = (stats_df['std'] / stats_df['mean']) * 100  # Coefficient of Variation

print("\nDescriptive Statistics:")
print(stats_df.round(2))

# ═══════════════════════════════════════════════════════════════════════════════
# 3. NUMERICAL FEATURES VISUALIZATION
# ═══════════════════════════════════════════════════════════════════════════════

# Visualization feature
for col in numerical_cols:
    fig = plt.figure(figsize=(18, 5))
    gs = fig.add_gridspec(1, 4, hspace=0.3, wspace=0.3)
    
    # Histogram with KDE
    ax1 = fig.add_subplot(gs[0, 0])
    ax1.hist(df[col].dropna(), bins=50, color='steelblue', alpha=0.7, edgecolor='black', density=True)
    df[col].dropna().plot(kind='kde', ax=ax1, color='red', linewidth=2)
    ax1.axvline(df[col].mean(), color='green', linestyle='--', linewidth=2, label=f'Mean: {df[col].mean():.2f}')
    ax1.axvline(df[col].median(), color='orange', linestyle='--', linewidth=2, label=f'Median: {df[col].median():.2f}')
    ax1.set_title(f'{col} - Distribution', fontsize=12, fontweight='bold')
    ax1.set_xlabel('Value', fontsize=10)
    ax1.set_ylabel('Density', fontsize=10)
    ax1.legend(fontsize=9)
    ax1.grid(alpha=0.3)
    
    # Boxplot
    ax2 = fig.add_subplot(gs[0, 1])
    bp = ax2.boxplot(df[col].dropna(), vert=True, patch_artist=True,
                     boxprops=dict(facecolor='lightcoral', alpha=0.7),
                     medianprops=dict(color='darkred', linewidth=2),
                     whiskerprops=dict(color='black', linewidth=1.5),
                     capprops=dict(color='black', linewidth=1.5),
                     flierprops=dict(marker='o', markerfacecolor='red', markersize=5, alpha=0.5))
    ax2.set_title(f'{col} - Boxplot', fontsize=12, fontweight='bold')
    ax2.set_ylabel('Value', fontsize=10)
    ax2.grid(axis='y', alpha=0.3)
    
    # Q-Q Plot
    ax3 = fig.add_subplot(gs[0, 2])
    stats.probplot(df[col].dropna(), dist="norm", plot=ax3)
    ax3.set_title(f'{col} - Q-Q Plot', fontsize=12, fontweight='bold')
    ax3.grid(alpha=0.3)
    
    # Violin Plot
    ax4 = fig.add_subplot(gs[0, 3])
    parts = ax4.violinplot([df[col].dropna()], positions=[0], showmeans=True, showmedians=True)
    for pc in parts['bodies']:
        pc.set_facecolor('lightgreen')
        pc.set_alpha(0.7)
    ax4.set_title(f'{col} - Violin Plot', fontsize=12, fontweight='bold')
    ax4.set_ylabel('Value', fontsize=10)
    ax4.set_xticks([])
    ax4.grid(axis='y', alpha=0.3)
    
    plt.suptitle(f'Univariate Analysis: {col}', fontsize=14, fontweight='bold', y=1.02)
    plt.savefig(f'../data/results/03_univariate_{col}.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print(f"\n{'='*80}")
    print(f"Feature: {col}")
    print(f"{'='*80}")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Mode: {df[col].mode()[0]:.2f}")
    print(f"Std Dev: {df[col].std():.2f}")
    print(f"Variance: {df[col].var():.2f}")
    print(f"Range: {df[col].max() - df[col].min():.2f}")
    print(f"IQR: {df[col].quantile(0.75) - df[col].quantile(0.25):.2f}")
    print(f"Skewness: {df[col].skew():.2f} ({'Right-skewed' if df[col].skew() > 0 else 'Left-skewed'})")
    print(f"Kurtosis: {df[col].kurtosis():.2f} ({'Leptokurtic' if df[col].kurtosis() > 0 else 'Platykurtic'})")

# ═══════════════════════════════════════════════════════════════════════════════
# 4. CATEGORICAL FEATURES ANALYSIS
# ═══════════════════════════════════════════════════════════════════════════════

categorical_cols = ['Gender', 'City', 'ContractType', 'PhoneService', 'MultipleLines',
                    'InternetService', 'OnlineSecurity', 'OnlineBackup', 'TechSupport',
                    'StreamingTV', 'StreamingMovies', 'PaymentMethod', 'PaperlessBilling']

print("\n" + "="*80)
print("CATEGORICAL FEATURES ANALYSIS")
print("="*80)

# Visualization categorical features
for col in categorical_cols:
    value_counts = df[col].value_counts()
    value_pcts = df[col].value_counts(normalize=True) * 100
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 5))
    
    # Bar plot
    axes[0].bar(value_counts.index, value_counts.values, color='teal', alpha=0.7, edgecolor='black')
    axes[0].set_title(f'{col} - Frequency Distribution', fontsize=12, fontweight='bold')
    axes[0].set_xlabel('Category', fontsize=10)
    axes[0].set_ylabel('Count', fontsize=10)
    axes[0].tick_params(axis='x', rotation=45)
    axes[0].grid(axis='y', alpha=0.3)
    
    # bar
    for i, (idx, val) in enumerate(value_counts.items()):
        axes[0].text(i, val, f'{val:,}\n({value_pcts.iloc[i]:.1f}%)', 
                    ha='center', va='bottom', fontsize=9, fontweight='bold')
    
    # Pie chart
    colors = sns.color_palette('Set3', len(value_counts))
    wedges, texts, autotexts = axes[1].pie(value_counts.values, labels=value_counts.index, 
                                            autopct='%1.1f%%', startangle=90, colors=colors,
                                            textprops={'fontsize': 10, 'fontweight': 'bold'})
    axes[1].set_title(f'{col} - Proportion', fontsize=12, fontweight='bold')
    
    plt.suptitle(f'Univariate Analysis: {col}', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig(f'../data/results/03_univariate_{col}.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print(f"\n{'='*80}")
    print(f"Feature: {col}")
    print(f"{'='*80}")
    print(f"Unique Values: {df[col].nunique()}")
    print(f"Most Frequent: {df[col].mode()[0]} ({value_pcts.iloc[0]:.2f}%)")
    print(f"\nValue Counts:")
    print(pd.DataFrame({'Count': value_counts, 'Percentage': value_pcts.round(2)}))

# ═══════════════════════════════════════════════════════════════════════════════
# 5. SUMMARY VISUALIZATION
# ═══════════════════════════════════════════════════════════════════════════════

# Summarization numerical features
fig, axes = plt.subplots(5, 2, figsize=(16, 20))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):
    axes[idx].hist(df[col].dropna(), bins=40, color='mediumpurple', alpha=0.7, edgecolor='black')
    axes[idx].axvline(df[col].mean(), color='red', linestyle='--', linewidth=2, label='Mean')
    axes[idx].axvline(df[col].median(), color='green', linestyle='--', linewidth=2, label='Median')
    axes[idx].set_title(f'{col}', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel('Value', fontsize=10)
    axes[idx].set_ylabel('Frequency', fontsize=10)
    axes[idx].legend(fontsize=9)
    axes[idx].grid(alpha=0.3)

plt.suptitle('All Numerical Features - Distribution Summary', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.savefig('../data/results/03_all_numerical_summary.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*80)
print("UNIVARIATE EDA COMPLETED")
print("="*80)
```
