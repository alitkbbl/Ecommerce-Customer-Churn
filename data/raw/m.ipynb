```python
import pandas as pd
import numpy as np

file_path = "customer_churn_dataset.csv"

# خواندن فایل
df = pd.read_csv(file_path)

# 1. اطلاعات کلی
info_df = pd.DataFrame({
    'ویژگی': ['تعداد سطرها', 'تعداد ستون‌ها', 'حجم حافظه (MB)'],
    'مقدار': [df.shape[0], df.shape[1], f"{df.memory_usage(deep=True).sum() / 1024**2:.2f}"]
})
print("اطلاعات کلی:")
print(info_df)
print("\n")

# 2. نمونه داده
print("نمونه داده:")
print(df.head())
print("\n")

# 3. انواع داده
types_df = pd.DataFrame({
    'ستون': df.columns,
    'نوع داده': df.dtypes.values
})
print("انواع داده:")
print(types_df)
print("\n")

# 4. مقادیر گمشده
missing_df = pd.DataFrame({
    'ستون': df.columns,
    'تعداد گمشده': df.isnull().sum().values,
    'درصد گمشده': (df.isnull().sum().values / len(df) * 100).round(2)
})
missing_df = missing_df[missing_df['تعداد گمشده'] > 0]
print("مقادیر گمشده:")
print(missing_df if len(missing_df) > 0 else "هیچ مقدار گمشده‌ای وجود ندارد")
print("\n")

# 5. آمار توصیفی
print("آمار توصیفی:")
print(df.describe())
```
